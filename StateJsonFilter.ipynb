## NY
Create filter to rip out bills that aren't signed

Passed bills are marked with a '4' in the progress array

NY's lla is one prior to 'DELIEVERED TO GOVERNOR'

```python
with open('NY.json') as file:
    variable_name = json.load(file)

### Pulls out all the passed legislation - this includes resolutions. Those will get filtered out 
### later

passed_bill_list = []
for bill in variable_name:
    bill_data = bill['bill']
    
    pass_test = bill['bill']['progress']
    
    for test in pass_test:
        
        if test['event'] == 4:
            
            passed_bill_list.append(bill)

### Now that I've isolated the lla as one prior to DTG, I need to employ my indexing trick
### To pull out the dates for the p_lla and lla. I also now need to take my data and start
### stuffing it into a csv file.

legis_data = []

for bill in passed_bill_list:

    bill_data = bill['bill']['bill_id']
           
    for x, event in enumerate(bill['bill']['history']):

        if event['action'].lower() == 'DELIVERED TO GOVERNOR'.lower():
            
            transfer_list = []
            
            ### the p_lla
#             print('----')
#             print(event['action'] + ' ' + (event['date']))

            ### Now for the lla
            lla = bill['bill']['history'][x-1]
#             print(lla['action'] + ' ' + lla['date'])
            
            ### Let's append EVERYTHING!!!
            
            transfer_list.append(bill_data)
            transfer_list.append(bill['bill']['session_id'])
            transfer_list.append(bill['bill']['session']['session_name'])
            transfer_list.append(bill['bill']['session']['session_title'])
            transfer_list.append(bill['bill']['session']['year_start'])
            transfer_list.append(bill['bill']['session']['year_end'])
            transfer_list.append(bill['bill']['url'])
            transfer_list.append(bill['bill']['state_link'])
            transfer_list.append(bill['bill']['status'])
            transfer_list.append(bill['bill']['state'])
            transfer_list.append(bill['bill']['state_id'])
            transfer_list.append(bill['bill']['bill_number'])
            transfer_list.append(bill['bill']['title'])
            transfer_list.append(event['action'])
            transfer_list.append(event['date'])
            transfer_list.append(lla['action'])
            transfer_list.append(lla['date'])
            
            legis_data.append(transfer_list)

 
         
df = pd.DataFrame(legis_data)

df.columns = ['legis_bill_id','session_id','session_name','session_title','year_start','year_end',
              'url','state_link','status','state','state_id','bill_number','title',\
              'p_lla_action','p_lla_date','lla_action','lla_date']

df.to_csv('ny_legis_data.csv',sep='|',index=False)
df.to_csv('ny_legis_data_BACKUP.csv',sep='|',index=False)

print('data_length ' + str(len(legis_data)))
```
